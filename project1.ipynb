## Data comparision b/w two excel files

```python
#reading source and target files
source=pd.read_excel('Source.xlsx')
target=pd.read_excel('Target.xlsx') 
#setting primaykey as index
source_index=source.set_index('FINANCIAL_TERM_SK')
target_index=target.set_index('FINANCIAL_TERM_SK')
#sorted the data w.r.t the primary key
source_sorted=source_index.sort_values(by=['FINANCIAL_TERM_SK'],ascending=True)
target_sorted=target_index.sort_values(by=['FINANCIAL_TERM_SK'],ascending=True)
#importing the DataFrame to new excel
source_sorted.to_excel('source_sorted.xlsx')
target_sorted.to_excel('target_sorted.xlsx')
#reading new excel files
source_1=pd.read_excel('source_sorted.xlsx')
target_1=pd.read_excel('target_sorted.xlsx')
#comparing
source_1==target_1
```
